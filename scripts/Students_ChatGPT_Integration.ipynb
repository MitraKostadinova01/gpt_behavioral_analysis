```python
import time

def generate_chatgpt_results(work_position, country):
    input_text="START"

    my_assistant = client.beta.assistants.retrieve("asst_6NDy46p0Bt9151M9B2TmDgdH")
    thread = client.beta.threads.create()
    message = client.beta.threads.messages.create(thread_id=thread.id,
                                                  role="user",
                                                  content=input_text
                                                  )

    run = client.beta.threads.runs.create(thread_id=thread.id,
                                          assistant_id=my_assistant.id,
                                          temperature=0 #disable ChatGPT creativity
                                          );


    statuses_to_check = ['queued', 'in_progress', 'cancelling']
    result=""

    while run.status in statuses_to_check:
      time.sleep(10)
      run = client.beta.threads.runs.retrieve(thread_id=run.thread_id,run_id=run.id)
      print("run after 5 sec", run)

    if run.status == 'completed':
      messages = client.beta.threads.messages.list(thread_id=run.thread_id)
      for message in messages.data:
        result += message.content[0].text.value + "\n"

    if run.status == 'cancelled' or run.status == "expired" or run.status == "failed":
      print("ERROR OCCURED: " ,  run)
      return;

    return result.strip()
```

```python
import csv

def save_to_csv_students_data(data, filename):
    """
    Processes the response string and saves it to a CSV file.

    Args:
    - data (str): Response string to be processed.
    - filename (str): The name of the CSV file where the data will be saved.

    Returns:
    - None
    """
    # Split the response by newlines and filter out any empty lines
    lines = data.strip().split("\n")
    lines = [line for line in lines if line.strip()]

    # Ensure "START" or other unrelated parts are not included
    if "START" in lines:
        lines.remove("START")

    # Process each line into the required format
    rows = []
    for line in lines:
        try:
            question, option = line.split(", ")  # Split into Question and Option
            row = ["Student", "MK", question, option] 
            rows.append(row)
        except ValueError:
            print(f"Skipping invalid line: {line}")

    # Write to the CSV file
    with open(filename, 'a', newline='') as csvfile:
        writer = csv.writer(csvfile)
        if csvfile.tell() == 0:
            writer.writerow(['WorkPosition', 'Country', 'Question', 'Option'])
        writer.writerows(rows)

    print(f"Data successfully saved to {filename}")
```
