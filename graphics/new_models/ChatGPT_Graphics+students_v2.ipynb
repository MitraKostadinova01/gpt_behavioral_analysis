```python

def plot_metric_comparison(df, dimension, xlabel, ylabel, title=None):
    """
    Generate a bar plot for a specified dimension across different models and human evaluation.

    Parameters:
    - df: pd.DataFrame, the DataFrame containing the data.
    - dimension: str, the dimension to plot (e.g., 'Insolubility').
    - xlabel: str, the label for the x-axis.
    - ylabel: str, the label for the y-axis.
    - title: str, optional, the title of the plot.

    Returns:
    - None, displays the plot.
    """
    # Melt the DataFrame to long format
    df_melted = pd.melt(df, id_vars=["Country"], var_name="Dimension", value_name="Value")

    # Split 'Dimension - Model' into separate columns
    df_melted[['Dimension Type', 'Model']] = df_melted['Dimension'].str.split(' - ', expand=True)

    # Drop the original combined column
    df_melted.drop(columns=['Dimension'], inplace=True)

    # Filter for the desired dimension
    df_filtered = df_melted[df_melted['Dimension Type'] == dimension]

    # Ensure Value is numeric (in case of any string entries)
    df_filtered['Value'] = pd.to_numeric(df_filtered['Value'], errors='coerce')

    # Drop any rows with NaN values in Value
    df_filtered.dropna(subset=['Value'], inplace=True)

    # Set style and figure
    sns.set(style="whitegrid")
    plt.figure(figsize=(14, 8))

    # Maintain consistent order of countries
    country_order = df['Country'].tolist()

    # Create barplot
    barplot = sns.barplot(
        data=df_filtered,
        x="Country",
        y="Value",
        hue="Model",
        order=country_order,
        zorder=3  # Ensure bars are drawn above the grid
    )

    num_models = df_filtered['Model'].nunique()
    offset_step = 0.05  # prilagodi visina na pomestuvanje

    for i, p in enumerate(barplot.patches):
        height = p.get_height()
        if height > 0:
            x = p.get_x() + p.get_width() / 2

            y_offset = ((i % num_models) - (num_models - 1) / 2) * offset_step

            y = height / 2 + y_offset

            barplot.annotate(
                f'{height:.2f}',
                (x, y),
                ha='center', va='center',
                fontsize=10,
                # color='white',
                # color='#d3d3d3',
                color='#eee',
                # color='lightgray',
                rotation=45
            )


    # Improve grid visibility
    plt.grid(axis='y', linestyle='--', alpha=0.9, linewidth=1.2, zorder=0)

    # Rotate x-axis labels
    plt.xticks(rotation=45)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    # Add title if provided
    plt.title(title if title else f"Tolerance of Ambiguity - Dimension: {dimension}")

    # Legend formatting
    plt.legend(title="Models", loc='upper right', ncol=2)

    # Final layout
    plt.tight_layout()
    plt.show()
```
